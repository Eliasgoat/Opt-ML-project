# Here are the plots for the MLP model

```python
import json
import os
from visualization import plot_test_loss_per_optimizer, print_final_accuracy_per_optimizer

def load_json(path_):
    with open(path_, "r") as f:
        return json.load(f)

def plot_mlp_alternative():
    files = {
        "CIFAR10": "data/merged_MLP_CIFAR10.json",
        "SVHN": "data/merged_MLP_SVHN.json"
    }

    for dataset, filepath in files.items():
        if not os.path.exists(filepath):
            print(f"âŒ File not found: {filepath}")
            continue

        print(f"\nðŸ“‚ Processing file: {filepath}")
        data = load_json(filepath)

        # === Test loss curve ===
        plot_test_loss_per_optimizer(
            results=data,
            title=f"Test Loss per Optimizer on {dataset} (MLP)",
            save_path=f"{dataset}_MLP_test_loss"
        )

        # === Print final accuracies ===
        print(f"\nðŸ“ˆ Final accuracy summary for {dataset} (MLP):")
        print_final_accuracy_per_optimizer(data)

plot_mlp_alternative()

```
